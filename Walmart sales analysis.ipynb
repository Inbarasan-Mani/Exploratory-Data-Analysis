## Sample Analysis of Store 1

```python
sns.set(font_scale = 1)
```

```python
sns.pairplot(store1)
plt.title('Relations of various parameters in Walmart Sales Dataset');
```

```python
sns.set_style('whitegrid')
plt.figure(figsize= (10,5))

sns.scatterplot(data= store1, x= 'CPI', y='Weekly_Sales',hue = 'Holiday_Flag', palette = 'Accent_r')

plt.title('REPORT - 1: Analysis of weekly sales of store 1');
```

##### The Consumer Price Index measures the average change in prices over time that consumers pay for a basket of goods and services.

##### CPI is the most widely used measure of inflation

* From the above chart, can conclude that the CPI and Holidays fixes the targets of weekly sales. The sales was clustered in the range of CPI (210, 212), and the revenue is stable at that point. And along the ranges of (214 - 221) the sales was distributed  throughout the range and it earns maximum revenue, 

```python
sns.set_style('whitegrid')
plt.figure(figsize = (10,7))
sns.scatterplot(data = store1, x= 'Weekly_Sales', y = 'Temperature', hue = 'Fuel_Price', palette = "summer")
plt.title('REPORT 2: Sales trend according to temperature and fuel price')
sns.set(font_scale = 2)
```

* The Sales of Store 1 according to weekly sales, temperature, fuel_price. The sales was narrowed down in high temperature(70-90), and scattered in normal temperature(40-60). And also shows that the sales was widely distributed over the sales range, whereas the sales was lower and cumulative over high temperature range

```python
store1.Unemployment.unique()
```

```python
plt.figure(figsize = (10,5))
sns.set(font_scale = 1)
sns.barplot(data = store1, x = 'Unemployment', y = 'Weekly_Sales')
plt.title('REPORT 3: Relation Between Weekly_Sales and Unemployment');
```

* The graph shows the details of relation between Unemployment and Weekly Sales. The trend clearly shows that whatever the Unemployment rates are there, peoples are always need of goods and the sales will be good through out the week

```python
plt.figure(figsize = (20,10))
sns.set(font_scale = 1)
sns.histplot(data = data, x = 'Weekly_Sales')
plt.title('REPORT 4: Weekly Sales Analysis Frequency');
```

* The sale was shown a increasing trend in low weekly sales, and rarely distributed in high sale segments. In the range of 500000 million dollars the sale was peaked at counts of 600+ and remaining shows as poison distribution

```python
plt.title('No of rows per year')
sns.countplot(x= pd.to_datetime(data.Date).dt.year);

year = pd.to_datetime(data.Date).dt.year
```

```python
plt.figure(figsize = (10,5))
sns.violinplot(data = data, x = pd.to_datetime(data.Date).dt.year , y = 'Weekly_Sales');
plt.xlabel('Year')
plt.title('REPORT 5: Yearly Analysis of Sales');
```

* The graph indicates that the sales on yearly basis, the dataframe gives the information of sales of years 2010, 2011, 2012. The face were at peak in 2010, 2011 and also represents higher revenue, whereas the sales in the year of 2012 have any outliers, means that sales doesn't reach the maximum value

```python
# performance on related with store and revenue
plt.figure(figsize = (20,10))
sns.barplot(data = data, x = 'Store', y = 'Weekly_Sales', hue = 'Holiday_Flag')
plt.title(' REPORT 6: Performance of Stores on Basis of Sales report');
```

* This breakdowns that the performance of store regarding sales report. The stores like 4,20, 13,14,etc.,that performance better so that we can rely on those high performance stores for high revenue. And low performed stores like 5,33,44, etc., are has to be improved according to customer's need and through feedback. And sales was also determined by Holiday's and Weekdays.
